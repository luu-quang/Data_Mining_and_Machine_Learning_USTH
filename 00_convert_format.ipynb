Why do we do this? 

- Loading a 12.7 million row CSV file takes almost 3 minutes and eats up gigabytes of RAM. 
By converting it to a 'Parquet' file, we compress the data into a columnar format. 
This means our modeling scripts will load the data in 6 seconds instead of 3 minutes, 
preventing our laptops from crashing during the Datathon.


```python
import pandas as pd
import time
import os
```

```python
print("=" * 50)
print("PHASE 1: PARQUET COMPRESSION")
print("=" * 50)

def optimize_and_save(csv_path, parquet_path):
    if not os.path.exists(csv_path):
        print(f"[ERROR] Cannot find {csv_path}")
        return
        
    start_time = time.time()
    print(f"Loading {csv_path}...")
    
    # Read CSV
    # low_memory=False prevents dtype inference issues
    df = pd.read_csv(csv_path, low_memory=False)
    
    print(f"Dataset shape: {df.shape}")
    
    # Save as Parquet using Snappy compression
    df.to_parquet(
        parquet_path,
        engine='pyarrow',
        compression='snappy'
    )
    
    elapsed = time.time() - start_time
    
    print(f"Success! Saved to:")
    print(parquet_path)
    print(f"Completed in {elapsed:.2f} seconds.\n")


# Dataset directory
DATASET_DIR = "/home/gq11/Documents/USTH/ML2/Project/dataset"

# Train dataset
optimize_and_save(
    os.path.join(DATASET_DIR, "train.csv"),
    os.path.join(DATASET_DIR, "train_optimized.parquet")
)

# Test dataset
optimize_and_save(
    os.path.join(DATASET_DIR, "test.csv"),
    os.path.join(DATASET_DIR, "test_optimized.parquet")
)
```
